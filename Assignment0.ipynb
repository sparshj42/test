```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('IITK_Courses_Report.csv')
```

```python
df.isnull()
```

```python
df.info()
```

```python
df.dropna(subset=['Student Name','Course'],inplace = True)
```

```python
df
```

```python
df = df.reset_index(drop = True)
df
```

```python
def gm(values):
    values = np.array(values)
    values = values[~np.isnan(values)]
    return np.prod(values) ** (1 / len(values))
```

```python
for col in ['Marks (%)','Attendance (%)']:
    for i in range(len(df)):
        if pd.isna(df.loc[i,col]):
            above = df.loc[i - 1, col] if i - 1 >= 0 else np.nan
            below = df.loc[i + 1, col] if i + 1 < len(df) else np.nan
            df.loc[i, col] = gm([above, below])
```

```python
df
```

```python
nullrows = df[df.isna().any(axis=1)]
nullrows
```

```python
df.to_csv('cleaned_IITK_Courses_Report.csv')
```
